```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch


dataset = pd.read_csv("hierarchical_clustering_dataset.csv")
X = dataset.iloc[:, 0:4].values  


plt.figure(figsize=(10, 5))
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.title("Dendrogram")
plt.xlabel("Samples")
plt.ylabel("Euclidean distances")
plt.show()

from sklearn.cluster import AgglomerativeClustering
hierarchy = AgglomerativeClustering(n_clusters= 5, metric = 'euclidean', linkage= 'ward')
y_hc = hierarchy.fit_predict(X)

from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)  # or X_scaled if you standardized
plt.figure(figsize=(8,5))
plt.scatter(X_pca[y_hc == 0, 0], X_pca[y_hc == 0, 1], s=60, c='red', label='Cluster 1')
plt.scatter(X_pca[y_hc == 1, 0], X_pca[y_hc == 1, 1], s=60, c='blue', label='Cluster 2')
plt.scatter(X_pca[y_hc == 2, 0], X_pca[y_hc == 2, 1], s=60, c='green', label='Cluster 3')
plt.scatter(X_pca[y_hc == 3, 0], X_pca[y_hc == 3, 1], s=60, c='cyan', label='Cluster 4')
plt.scatter(X_pca[y_hc == 4, 0], X_pca[y_hc == 4, 1], s=60, c='magenta', label='Cluster 5')
plt.title('Hierarchical Clustering (PCA 2D Projection)')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.tight_layout()
plt.show()
```
